## Data Analysis

###  Top 5 most popular artists

```python
top_five_artist = df.groupby("artist").count().sort_values(by = "song_title", ascending=False)["song_title"][:5]
top_five_artist
```

```python
top_five_artist.plot.barh()
```

### Top 5 loudest track

```python
top_five_loudest_tracks = df[["loudness","song_title"]].sort_values(by= "loudness", ascending =True)[:5]
top_five_loudest_tracks
```

#### visualize

```python
plt.figure(figsize=(12,7))
sns.barplot(x="loudness", y = "song_title", data = top_five_loudest_tracks)
plt.title("Top 5 loudest track")
plt.show()
```

##  Artist with the most danceable songs

```python
top_ten_artist_with_dancable_song = df[["danceability", "song_title", "artist"]].sort_values(by = "danceability", ascending = False)[:10]
top_ten_artist_with_dancable_song
```

#### visualize

```python
plt.figure(figsize=(12,7))
sns.barplot(x ="danceability", y="artist", data = top_ten_artist_with_dancable_song)
plt.title = "Artist with most danceable song"
plt.show()
```

## Top 5 instrumental songs

```python
top_five_instrumental_songs = df[["instrumentalness", "song_title", "artist"]].sort_values(by = "instrumentalness", ascending = False)[:5]
top_five_instrumental_songs
```

#### visualize

```python
plt.figure(figsize = (12,7))
plt.pie(x ="instrumentalness", data = top_five_instrumental_songs, autopct = "%1.2f%%", labels = top_five_instrumental_songs.song_title)
plt.suptitle("Top 5 instrumentalness song")
plt.show()
```

## Multiple feature plots

```python
interest_feature_cols = ["tempo", "loudness", "acousticness", "danceability", "duration_ms", 
                         "energy", "instrumentalness", "liveness", "speechiness", "valence"]
```

```python
for feature_col in interest_feature_cols:
    
    pos_data = df[df["target"]==1][feature_col]
    neg_data = df[df["target"]==0][feature_col]
    
    plt.figure(figsize = (12,7))
    
    sns.distplot(pos_data, bins =30, label = "Positive", color = "green")
    sns.distplot(neg_data, bins =30, label = "Negative", color = "red")
    
    plt.legend(loc="upper right")
    plt.suptitle(f"Positive and Negative plot for {feature_col}")
    plt.show() 
```

## Top 10 energetic tracks

```python
top_ten_energetic_songs = df[["energy", "song_title", "artist"]].sort_values(by = "energy", ascending = False)[:10]
top_ten_energetic_songs
```

#### Visualize

```python
plt.figure(figsize = (12,7))
sns.barplot(x="song_title", y ="energy", data = top_ten_energetic_songs)
plt.suptitle = "Top ten energetic song"
```

```python
plt.figure(figsize = (12,7))
plt.pie(x="energy", data = top_ten_energetic_songs, autopct = '%1.2f%%')
plt.suptitle = "Top ten energetic song"
```
